```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/train_delay.csv")
df.head()
```

```python
y = df['delay_minutes']
```

```python
X = df.drop(columns=['delay_minutes', 'train_number'])
```

Train number was excluded due to high cardinality and lack of predictive pattern

```python
X.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_cols = ['train_type', 'source', 'destination', 'day_of_week']
numerical_cols = ['distance_km', 'month']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols),
        ('num', 'passthrough', numerical_cols)
    ]
)
```

```python
X_transformed = preprocessor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_transformed,
    y,
    test_size=0.2,
    random_state=42
)
```

Data was split into 80% training and 20% testing to evaluate model performance
